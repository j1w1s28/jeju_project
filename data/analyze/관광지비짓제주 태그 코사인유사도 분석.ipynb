```python
title_to_index = dict(zip(visit_tour['title'], visit_tour.index))

# 영화 제목 Father of the Bride Part II의 인덱스를 리턴
idx = title_to_index['협재해수욕장']
print(idx)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim2):
    # 선택한 영화의 타이틀로부터 해당 영화의 인덱스를 받아온다.
    idx = title_to_index[title]

    # 해당 관광지와 모든 관광지와의 유사도를 가져온다.
    sim_scores = list(enumerate(cosine_sim[idx]))

    # 유사도에 따라 관광지들을 정렬한다.
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # 가장 유사한 10개의 관광지를 받아온다.
    sim_scores = sim_scores[1:11]

    # 가장 유사한 10개의 관광지의 인덱스를 얻는다.
    tour_indices = [idx[0] for idx in sim_scores]

    # 가장 유사한 관광지의 장소이름을 리턴한다.
    return visit_tour['title'].iloc[tour_indices]
```
